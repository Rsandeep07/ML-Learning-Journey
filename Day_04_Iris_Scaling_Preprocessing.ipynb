```python
def order_data(x):
  if x > 4:
    return 'L'
  elif x > 2:
    return 'M'
  else:
    return 'S'
```

```python
tips['size_cat'] = tips['size'].apply(order_data)
tips.head()
```

```python
tips['size'].apply(order_data).unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder().set_output(transform = 'pandas')
oe
```

```python
oe.fit_transform(tips[['size_cat']])
```

```python
order = ['S', 'M', 'L']
```

```python
oe = OrdinalEncoder(categories = [order]).set_output(transform = 'pandas')
oe.fit_transform(tips[['size_cat']])
```

```python
ohe = OneHotEncoder(sparse_output= False, drop = 'first')
oe = OrdinalEncoder(categories=[order])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer([('OHE', ohe, ['sex', 'smoker', 'day', 'time']),
                        ('OE', oe, ['size_cat'])],
                       remainder = 'passthrough').set_output(transform = 'pandas')
ct
```

```python
ct.fit_transform(tips[:])
```
