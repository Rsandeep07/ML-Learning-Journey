# IRIS

### LOAD DATA

```python
from sklearn.datasets import load_iris
```

```python
# object
data = load_iris()
```

```python
data
```

```python
data.keys()
```

```python
features = data.data
features
```

```python
feature_names = data.feature_names
feature_names
```

```python
target = data.target
target
```

```python
target_names = data.target_names
target_names
```

```python
import numpy as np
import pandas as pd
```

```python
iris_data = pd.DataFrame(features,columns=feature_names)
```

```python
iris_data
```

```python
iris_data['target'] = target
iris_data.head()
```

```python
iris_data.target.unique()
```
