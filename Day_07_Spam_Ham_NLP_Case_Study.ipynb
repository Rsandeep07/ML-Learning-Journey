- **Upsampling:** Increasing the number of samples in the minority class (e.g., by duplicating or generating synthetic data) so both classes have similar sizes.

- **Downsampling:** Reducing the number of samples in the majority class (e.g., by randomly removing data) to balance the dataset.

- Both samples are used to handle the imbalanced datasets so the model doesnâ€™t get biased toward one class.
