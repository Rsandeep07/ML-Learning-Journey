## 1Ô∏è‚É£ Recap: Error in Machine Learning

In Machine Learning, our objective is to minimize prediction error.

\[
\text{Error} = y - \hat{y}
\]

Where:
- **y** ‚Üí Actual value  
- **≈∑** ‚Üí Predicted value  

A good model ensures:
\[
\hat{y} \approx y
\]

## 2Ô∏è‚É£ Bias

### üîπ What is Bias?
Bias occurs when a model is **too simple** and makes **strong assumptions** about the data.

Examples:
- Mean model  
- Mode model  
- Very simple linear models  

### üîπ Why Bias Happens?
- Model is built on simple assumptions  
- Very little scope for learning  

### üîπ Effect of High Bias
- Training error ‚Üí High  
- Testing error ‚Üí High  

‚û°Ô∏è Leads to **Underfitting**

---

## 3Ô∏è‚É£ Variance

### üîπ What is Variance?
Variance refers to **variation in model performance** when training data changes slightly.

### üîπ Why Variance Happens?
- Model is very complex  
- Learns noise and random patterns  
- Highly sensitive to training data  

### üîπ Effect of High Variance
- Training performance ‚Üí Very high  
- Testing performance ‚Üí Low  

‚û°Ô∏è Leads to **Overfitting**

---

## 4Ô∏è‚É£ Bias‚ÄìVariance Trade-off

| Model Type | Bias | Variance | Result |
|-----------|------|----------|--------|
| Simple Model | High | Low | Underfitting |
| Complex Model | Low | High | Overfitting |

üéØ Goal: Balance bias and variance for good generalization.

---

## 5Ô∏è‚É£ Detecting Underfitting & Overfitting

### üîç Method:
Compare **Training Performance** vs **Testing Performance**

### ‚úîÔ∏è Underfitting:
- Train performance ‚Üí Low  
- Test performance ‚Üí Low  

### ‚ùå Overfitting:
- Train performance ‚Üí High  
- Test performance ‚Üí Low  

---

## 6Ô∏è‚É£ Why Not Tune Using Test Data?

- Test data must be used **only once**
- Repeated evaluation causes **data leakage**
- Model indirectly learns test patterns

üö´ Hence, tuning with test data is **not recommended**

---

## 7Ô∏è‚É£ Validation Data ‚Äì Correct Approach

### üîπ Dataset Split


```text
Dataset
‚îú‚îÄ‚îÄ Training Data
‚îÇ ‚îú‚îÄ‚îÄ Training Set
‚îÇ ‚îî‚îÄ‚îÄ Validation Set
‚îî‚îÄ‚îÄ Test Data


### üîπ Purpose of Validation Data
- Tune model parameters  
- Detect generalization  
- Avoid data leakage  

---

## 8Ô∏è‚É£ Cross Validation ‚Äì Core Concept

Cross Validation is used when:
- Dataset is limited  
- Models are computationally affordable  

### üîÅ Process:
1. Divide training data into **K parts**
2. Use **K-1 parts for training**
3. Use **1 part for validation**
4. Repeat **K times**
5. Average validation performance

---

## 9Ô∏è‚É£ Types of Cross Validation

1. **K-Fold Cross Validation**  
2. **Repeated K-Fold Cross Validation**  
3. **Stratified K-Fold (for classification)**  
4. **Leave-One-Out Cross Validation (LOOCV)**  
5. **Leave-P-Out Cross Validation (LPOCV)**  

(Available in **scikit-learn**)

---

## üîü Detecting Generalization Using Cross Validation

- If mean(train score) ‚âà mean(validation score)  
  ‚Üí Model generalizes well  

- If train score >> validation score  
  ‚Üí Overfitting  

---

## 1Ô∏è‚É£1Ô∏è‚É£ Improving Generalization

### üîπ Regularization (Parametric Models)
- L1 (Lasso)
- L2 (Ridge)
- Elastic Net

### üîπ Ensemble Methods
- **Bagging** ‚Üí Reduces variance  
- **Boosting** ‚Üí Reduces bias  

---

## 1Ô∏è‚É£2Ô∏è‚É£ Within-Subject vs Between-Subject Design

### üîπ Between-Subject Design
- Each observation appears only once  
- Different subjects for different conditions  

### üîπ Within-Subject Design
- Same subject observed multiple times  
- Common in healthcare & social science studies  

‚ö†Ô∏è Requires careful validation to avoid data leakage.

---

## 1Ô∏è‚É£3Ô∏è‚É£ Key Takeaways

- Bias causes underfitting  
- Variance causes overfitting  
- Never tune using test data  
- Use validation & cross validation  
- Cross validation improves generalization  
- Proper data design is crucial in real-world ML

---
